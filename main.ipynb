## 2. Now write a function that returns the largest element on a list

```python
def greatest (elements):
    """
    This function takes a list of numbers and returns the maximum value.
    If the list is empty, it returns None.
    """
    
    max_value = elements[0]
    for number in elements:
        if number > max_value:
            max_value = number
    return max_value
      
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here

  print(n, '!') # this is just to show the recursion in action
  # corner case: 1! = 1
  if n <= 0:  #this is the base case for the recursion and where the recursion stops
    return 1 #this is the base return value for the recursion to work correctly

  # Recursive case: n! = n * (n-1)!
  else:
      local_factorial = n * factorial(n-1)  #this is the recursive call
      #print("Factorial for " + str(n) +"! is " + str(local_factorial))
      return local_factorial

factorial(5)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python

#your code here

"""This code calculates the standard deviation of a list"""

def st_dev(list_sd):
    count = 0 
    for _ in list_sd:
        count += 1

    # here is where the mean is being calculated
    total = 0
    for num in list_sd:
        total += num
    mean = total / count

    # here is where the squared differences are being calculated from the mean
    sum_squared_diffs = 0
    for num in list_sd:
        diff = num - mean
        sum_squared_diffs += diff * diff

    #calculating the variance
    variance = sum_squared_diffs / (count - 1)
   
    return variance ** 0.5

```

```python
# This will test your function 
test_stdev(st_dev)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    #your code here

    if len(password) < 8:
        return False 

    check1 = 0 #checks the number requirement
    check2 = 0 #checks lower case requirement
    check3 = 0 #checks upper case requirement
    check4 = 0 #checks special character

    for char in password:
        if "0" <= char <= "9":
            check1 = 1
        elif "a" <= char <= "z":
            check2 = 1
        elif "A" <= char <= "Z":
            check3 = 1
        else:
            check4 = 1

    if check1 == 1 and check2 == 1 and check3 == 1 and check4 == 1:
        return True
    else:
        return False



```
